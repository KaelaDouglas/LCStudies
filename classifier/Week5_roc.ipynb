```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_all = np.load('roc_all.npz')
f_b0715 = np.load('roc_b0715.npz')
f_b1425 = np.load('roc_b1425.npz')
f_b2531 = np.load('roc_b2531.npz')
f_g15 = np.load('roc_g15.npz')
f_l07 = np.load('roc_l07.npz')
f_l05 = np.load('roc_l05.npz')
f_l1 = np.load('roc_l1.npz')
f_l15 = np.load('roc_l15.npz')
```

```python
fp_all, tp_all, auc_all = f_all['arr_0'], f_all['arr_1'], f_all['arr_2']
fp_b0715, tp_b0715, auc_b0715 = f_b0715['arr_0'], f_b0715['arr_1'], f_b0715['arr_2']
fp_b1425, tp_b1425, auc_b1425 = f_b1425['arr_0'], f_b1425['arr_1'], f_b1425['arr_2']
fp_b2531, tp_b2531, auc_b2531 = f_b2531['arr_0'], f_b2531['arr_1'], f_b2531['arr_2']
fp_g15, tp_g15, auc_g15 = f_g15['arr_0'], f_g15['arr_1'], f_g15['arr_2']
fp_l07, tp_l07, auc_l07 = f_l07['arr_0'], f_l07['arr_1'], f_l07['arr_2']
fp_l05, tp_l05, auc_l05 = f_l05['arr_0'], f_l05['arr_1'], f_l05['arr_2']
fp_l1, tp_l1, auc_l1 = f_l1['arr_0'], f_l1['arr_1'], f_l1['arr_2']
fp_l15, tp_l15, auc_l15 = f_l15['arr_0'], f_l15['arr_1'], f_l15['arr_2']
```

```python
plt.figure(figsize=[8,8])
plt.ylim(-.01, 1.01)
plt.xlim(-.01, 1.01)
plt.plot(fp_all, tp_all, label='all $\eta$, AUC = %.3f'%auc_all, linewidth=3, c='xkcd:dirt brown')
plt.plot(fp_b0715, tp_b0715, label='.7 < $\eta$ < 1.5, AUC = %.3f'%auc_b0715, linewidth=3, c='xkcd:dark tan')
plt.plot(fp_b1425, tp_b1425, label='1.4 < $\eta$ < 2.5, AUC = %.3f'%auc_b1425, linewidth=3, c='xkcd:caramel')
plt.plot(fp_b2531, tp_b2531, label='2.5 < $\eta$ < 3.1, AUC = %.3f'%auc_b2531, linewidth=3, c='xkcd:fawn')
plt.plot(fp_g15, tp_g15, label='$\eta$ > 1.5, AUC = %.3f'%auc_g15, linewidth=3, c='xkcd:brown')
plt.plot(fp_l07, tp_l07, label='$\eta$ < .7, AUC = %.3f'%auc_l07, linewidth=3, c='xkcd:dark brown')
plt.plot(fp_l05, tp_l05, label='$\eta$ < .5, AUC = %.3f'%auc_l05, linewidth=3, c='xkcd:butterscotch')
plt.plot(fp_l1, tp_l1, label='$\eta$ < 1., AUC = %.3f'%auc_l1, linewidth=3, c='xkcd:light peach')
plt.plot(fp_l15, tp_l15, label='$\eta$ < 1.5, AUC = %.3f'%auc_l15, linewidth=3, c='xkcd:pumpkin')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.legend()
plt.savefig('./pfn_eta_rocs')
```
